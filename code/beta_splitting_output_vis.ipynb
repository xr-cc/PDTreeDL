```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## Load output data

```python
beta_classes = ['beta_1_yule', 'beta_2_pda', 'beta_3_ab', 'beta_4_balanced', 'beta_5_unbalanced']
beta_values = [int(item.split("_")[1]) for item in beta_classes]

FILE_PATH = "../output/beta-splitting/"
df_train_latent = pd.read_csv(FILE_PATH+"beta_splitting_training_latent.csv")
df_test_latent = pd.read_csv(FILE_PATH+"beta_splitting_test_latent.csv")
df_clf = pd.read_csv(FILE_PATH+"beta_splitting_clf_results.csv")
```

```python
df_train_latent["class"] = df_train_latent["class"].str.split("_",expand=True)[1].apply(pd.to_numeric)
df_test_latent["class"] = df_test_latent["class"].str.split("_",expand=True)[1].apply(pd.to_numeric)
```

## Plot results

```python
fig, axes = plt.subplots(3,1,sharex=False, sharey=False, figsize=(5,14), dpi=80)

## latent space of training data
ax = axes[0]

sns.scatterplot(ax=ax, data=df_train_latent, x='LF1', y='LF2', hue='class', 
                palette="muted", alpha=0.5)
ax.set_xlabel('LF1', fontsize=13)
ax.set_ylabel('LF2', fontsize=13)
ax.set_title("Training Data", fontsize=14)

ax.set_xlim(-4.5,4.5)
ax.set_ylim(-3,3)
ax.set_aspect(1) 

ax.legend(loc='upper left',prop={'size':11}, title='beta',title_fontsize=12)

## latent space of test data
ax = axes[1]

sns.scatterplot(ax=ax, data=df_test_latent, x='LF1', y='LF2', hue='class',
                palette="muted", alpha=0.5)
ax.set_xlabel('LF1', fontsize=13)
ax.set_ylabel('LF2', fontsize=13)
ax.set_title("Test Data", fontsize=14)

ax.set_xlim(-4.5,4.5)
ax.set_ylim(-3,3)
ax.set_aspect(1) 
ax.get_legend().remove()

## confusion chart
ax = axes[2]

sns.set(font_scale=1.2)
cm = confusion_matrix(df_clf[df_clf['type']=='test']['true'], df_clf[df_clf['type']=='test']['pred']) #,normalize='true')
df_cm = pd.DataFrame(cm,index=beta_values,columns=beta_values)
g = sns.heatmap(df_cm, ax=ax, linewidths=0.5,
                cbar_kws={"ticks":np.arange(0,201,50),"shrink":0.85, "label": 'Number of samples'},
                annot=True,fmt="d",annot_kws={"fontsize":13}, cmap="YlGnBu",square=True,)
# ax.set_title("Confusion Plot", fontsize=14)
g.set(xlabel='Predicted beta', ylabel='True beta')
plt.subplots_adjust(wspace=0.12,  hspace=0.12)

# save
plt.savefig("../plots/beta_splitting.pdf", bbox_inches='tight', format="pdf", dpi=50)
```
